```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
import matplotlib.colors as colors
import matplotlib.cm as cmx

%matplotlib inline
%config InlineBackend.figure_format = 'retina'

import seaborn
seaborn.set(style='whitegrid')
seaborn.set_context('notebook')
```

```python
df = pd.read_csv("data/DataFrame_20gen_10inds_5objects.csv")
new_columns = df.columns.values
new_columns[0] = 'gen'
df.columns = new_columns
df = df.set_index("gen")
df.head()
```

```python
df.plot(title="Optimization of the Tweaker")
plt.ylim((0, df["median"].max()*1.1))
plt.hlines(df["top"].min(), xmin=1, xmax=df.index.max())
plt.figtext(.7, .5, f"minimal error: \n{df['top'].min().round(6)}")
plt.show()
```
